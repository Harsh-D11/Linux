8. You suspect that a particular process is consuming excessive CPU resources on your Linux server. How
would you identify and terminate this process?
>- - 1.  Spotting the Culprit:
Your first move should be to open a terminal and run top (or the more user-friendly htop, if you have it installed). These tools show you a live list of processes, usually sorted by how much CPU they're using, putting the heaviest users right at the top. Note down the PID (Process ID) of the suspicious process.

  - - 1.1 - (A quick command-line way to list the top few CPU users: ps aux --sort=-%cpu | head -n 5)

 - - 2.Taking Action (Terminating it):
Once you have the PID, you use the kill command.

 - - 2.1 - Gentle nudge: First, try kill <PID> (e.g., kill 12345). This sends a signal asking the process to shut down gracefully, allowing it to clean up.

 - - 2.2 - Forceful shutdown: If it's stubborn and doesn't close, you might need to use kill -9 <PID> (e.g., kill -9 12345). This is like pulling the plug; it stops the process immediately, but it won't do any cleanup, so use it only if the graceful method fails.
